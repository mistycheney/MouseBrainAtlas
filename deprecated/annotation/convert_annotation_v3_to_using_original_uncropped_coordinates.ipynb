```python
# Map original to aligned-cropped

for cnt_id, cnt in contours.iterrows():
    fn = cnt['filename']
    sec = filename_to_section[fn]
    contours.loc[cnt_id, 'section'] = sec

    T = Ts[fn].copy()
    T[:2, 2] = T[:2, 2]*32
    Tinv = np.linalg.inv(T)

    print cnt['vertices'].dtype
    
    n = len(cnt['vertices'])
    a = np.dot(Tinv, np.c_[cnt['vertices'], np.ones((n,))].T).T[:, :2] - (xmin*32, ymin*32)
    print a.dtype

    contours.set_value(cnt_id, 'vertices', a)

#     label_position_on_aligned_cropped = np.dot(Tinv, np.r_[cnt['label_position'], 1])[:2] - (xmin*32, ymin*32)
#     contours.loc[cnt_id, 'label_position'] = label_position_on_aligned_cropped.tolist()
```

```python
# Map aligned-cropped to original

for cnt_id, cnt in contours.iterrows():
    
    sec = cnt['section']
    fn = sorted_filelist[sec] 
    contours.loc[cnt_id, 'filename'] = fn
    
    T = Ts[fn].copy()
    T[:2, 2] = T[:2, 2]*32
#     Tinv = np.linalg.inv(T)
    
    n = len(cnt['vertices'])
    
    vertices_on_aligned = np.array(cnt['vertices']) + (xmin*32, ymin*32)    
    contours.loc[cnt_id, 'vertices'] = np.dot(T, np.c_[vertices_on_aligned, np.ones((n,))].T).T[:, :2]
#     contours.loc[cnt_id]['vertices'] = 0
    
    label_position_on_aligned = np.array(cnt['label_position']) + (xmin*32, ymin*32)
    contours.loc[cnt_id, 'label_position'] = np.dot(T, np.r_[label_position_on_aligned, 1])[:2]

del contours['section']
```
